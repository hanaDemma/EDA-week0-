```python
# Check for negative values in specific columns
columns_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']
for col in columns_to_check:
    print(f"{col} has negative values: {data[col][data[col] < 0].count()}")
```

```python
# Summary statistics for identifying outliers
summary_stats = data[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']].describe()
```

```python
# Detecting outliers using the IQR method for selected columns
def detect_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((column < lower_bound) | (column > upper_bound)).sum()
    return outliers

outliers = {
    column: detect_outliers_iqr(data[column])
    for column in ['ModA', 'ModB', 'WS', 'WSgust']
}
print("\nSummary Statistics:\n", summary_stats)
print("\nOutliers Detected (IQR Method):\n", outliers)
```

```python
# Ensure the 'Timestamp' column is a datetime object
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
# Extract time-based features
data['Month'] = data['Timestamp'].dt.month
data['Hour'] = data['Timestamp'].dt.hour
```

```python
# Aggregate data by Month and Hour
monthly_avg = data.groupby('Month')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()
hourly_avg = data.groupby('Hour')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()
```

```python
# Plotting monthly trends
plt.figure(figsize=(14, 8))

plt.subplot(2, 1, 1)
monthly_avg.plot(kind='bar', ax=plt.gca(), colormap='viridis', alpha=0.8)
plt.title("Monthly Averages of GHI, DNI, DHI, and Tamb")
plt.ylabel("Values")
plt.xlabel("Month")
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.legend(loc='upper left')
```

```python
# Plotting hourly trends
plt.subplot(2, 1, 2)
hourly_avg.plot(ax=plt.gca(), colormap='coolwarm', alpha=0.8)
plt.title("Hourly Averages of GHI, DNI, DHI, and Tamb")
plt.ylabel("Values")
plt.xlabel("Hour")
plt.xticks(ticks=range(0, 24, 2))
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
